```python
#NAME: Busra Karaoglan, DATE: 21.12.2018, PRID:    , REG.NO:    










#import necessary packages for following functions
import csv
import math
import random



# This function have filename as the only paramater. 
# It read a CSV file (this is why we import csv at the beginning)
# and return a matrix in which a row in the csv-file 
# is a row in the matrix. 
# Because CSV files have comma separates columns, we need to use split(',').
# And last all the values have to be converted from string to int.

def load_from_csv(filename):
    with open(filename) as file:
        matrix=[]
        for line in file:
            line=line.strip().split(',')                       
            int_list = [int(x) for x in line]  #converting a string to int        
            matrix.append(int_list)
    return matrix





# This function have two list parameters.
# It return the Euclidean distance between the this two lists. 
# The details about the Euclidean distance can be read at the appendix of the Coursework PDF file.




# This function is needed to !!!!!!!!!!!!!!!!



def get_distance(list_a, list_b):
    total = 0
    
    for i in range(len(list_a)):
        total+=(list_a[i]-list_b[i])**2
    dist=total**0.5
    
    return dist





# This function is implemented because we need to use it in the following function.
# It has two parameters, x is a matrix and j is a column number.
# It returns the standard deviation of the elements in the column number passed as a parameter.
# The details about the calculation of standard deviation is written in the Coursework appendix.

def get_standard_devitation(x,j):
    
    total = 0
    num_rows=len(x)
    
    
    #sum the values of a row
    for i in range(num_rows):
        total+=x[i][j]
    
    #calculate the mean of them
    mean=total/num_rows    
    
    total = 0            
    for i in range(num_rows):
        total+=(x[i][j]-mean)**2
            
    div = (total)/(num_rows-1)
    
    std = math.sqrt(div)

    return std









# This function has just one parameter, which is a matrix (here the variable x).
# This function use the previous get_standard_deviation function to standardise the matrix x.
# That means, the matrix x which is passed as a parameter to this function,
# return a matrix containing the standardised version of x.
# The details for this calculation can be read again at the Coursework appendix.

def get_standardised_matrix(x):
    
    # declare this variables, so we can use this simpler versions in loops
    num_rows=len(x)
    num_columns = len(x[0])
    
    # both of them is needed later, so we initialize them here to be zero and empty.
    total = 0
    meanlist=[]
    
    for j in range(num_columns):
        for i in range(num_rows):
            total+=x[i][j]
            mean=total/num_rows
        total=0  # total back to =0, because otherwise it would be overwritten in the j loop.
        meanlist.append(mean)

    
    # declare an empty matrix and an empy array1 to append the new standardised values of matrix x
    matrix = [[0]*num_columns]*num_rows
    array1=[]
    
    for i in range(num_rows):
        for j in range(num_columns):
            matrix[i][j] = (x[i][j]-meanlist[j])/(get_standard_devitation(x,j))
            array1.append(matrix[i][j])
    
    
    # we get from array1 a list with the values for the standardises matrix,
    # but we need this as matrix - one way to do so is to split this array1 in to list of list = matrix
    # therefore, here is needed another list(array2) which would be at the end of the while loop a matrix.
    array2 = []
    while len(array1) > num_columns:
        pices = array1[:num_columns]
        array2.append(pices)
        array1 = array1[num_columns:]
        
    array2.append(array1)
    
    return array2
    




# The following function have four parameters: 
# One of them is a list (here data_row). 
# It is a row of the test data matrix which we going to use later in the run_test function.
# The other two parameters are matrix. The first one is called learn_x and it contain the data from the file learning_data
# and the other one is called learn_y and it contain the data of the file learning_data_labels.
# The last parameter is a positive integer k. 



# This function is the k-nearest neighbors (or k-NN for short) and is a simple machine learning algorithm.
# It categorizes an input by using its k nearest neighbors.

# Summarized, this function find the k rows of the matrix learning_data 
# that are the closest to the list passed as a parameter. 
# Here, we use the get_distance function to identify this rows which are the closest.
# After finding these k rows, it return the related rows in the matrix learning_data_labels.
   
    
    
    
    
def get_k_nearest_labels(data_row, learn_x, learn_y, k):
    
    # we need empty distance list, so we can append the euclidian distances which
    # we going to find with the help of get_distance.
    distance = []
    
    for i in range(len(learn_x)): 
        new = [get_distance(data_row, learn_x[i]),i] 
        # with ,i we find the row numbers in which the distance is calculated
        distance.append(new)
    
    # with .sort() the distance list is sorted from the minimum to the maximum distance
    distance.sort()
    
    # the nearest list is going to be filled with the k minimum distance thank to following while loop
    nearest = []
 
    i= 0
    while i<k:
        nearest.append(distance[i])
        i+=1
    
    # now we want to know the rows of this k minimum distances - the row numbers from the k nearest neigbours
    labels_row = []
    
    for i in range(len(nearest)):
        labels_row.append(nearest[i][1])
    
    
    # lastly, we return the the labels of this k nearest neigbours
    labels = []
    
    for i in range(len(labels_row)):
        labels.append(learn_y[labels_row[i]])
    
    return labels






# It can happen that the k nearest neigbour labels are not all the same. 
# Therefore, we should take this label which is the most common. 
# This function get_mode with just one parameter labels will do this.

# The matrix parameter labels is returned by get_k_nearest_labels 
# and have therefore a matrix just only one column.

# This function return the mode of the numbers in this labels matrix. 
# The mode of a sequence of numbers is the number with the highest frequency.
# If more than one number has the highest frequency, then it return one 
# of the numbers with the highest frequency at random (therefore, import random).

def get_mode(labels):
     
    
    # we are going to get a matrix a parameter - but it would be preferable to get a list
    # therefore, this matrix is going to be flattened in the following for loop
    flattened_list = []

    for i in labels:
        for j in i:
            flattened_list.append(j)
    
    labels = flattened_list
    
    
    # now we use dictionary as a counter
    count_dict = {}
    
    for i in (labels):   
        count = labels.count(i)        
        if i not in count_dict.keys():
            count_dict[i] = count
    
    max_count = 0 
    
    for key in count_dict: 
        if count_dict[key] >= max_count:
            max_count = count_dict[key]
    
    corr_keys = [] 
    
    for corr_key, count_value in count_dict.items():
        if count_dict[corr_key] == max_count:
            corr_keys.append(corr_key)
    
    # if one or more numbers are occured at the same amount, than we choose randomly one of them 
    corr_keys = sorted(corr_keys)
    return random.choice(corr_keys)




# The classify function have 4 parameters:
# testX is the the matrixes for data in which we get later our testing matrix.
# trainX is the matrix from learning_data,
# trainY is the matrix with the labels from the learning_data_labels
# and k is the positive intger which we will use to get the k nearest neighbours.


# First of all, the matrixses for testX and trainX have to be standardised beforehand,
# so the algorithm of k nearest neigbours work correctly.



#This function follow the algorithm described in the appendix. It should return a matrix; in this
#document we call this matrix data_labels.
#This function should use the other functions you wrote as much as possible. Do not keep repeating
#code you already wrote.



def classify(testX, trainX, trainY, k):
    
    #first standardise the matrixes testX and trainX
    testX = get_standardised_matrix(testX)
    trainX = get_standardised_matrix(trainX)
    
    #hier train und test X stan oder nicht bei k neighbours???
    
    classifying = []
    
    for row in testX:
        liste = get_k_nearest_labels(row, trainX, trainY, k)
        label = get_mode(liste)
        classifying.append(label)
    return classifying





def get_accuracy(correct_label, algo_label):
    
    acc = 0
    count = 0
    for i in range(len(algo_label)):
        if algo_label[i] == correct_label[i][0]: #hier noch bei correct label also bei testY [0] da das matrix ist und nur 1 spalte hat
            count+=1
    
    acc = (100/len(algo_label))*count
    
    return acc





def run_test(testX, testY, trainX, trainY):
    
    # die files reinladen
    testX = load_from_csv(testX)
    testY = load_from_csv(testY)
    trainX = load_from_csv(trainX)
    trainY = load_from_csv(trainY)
    
    # stadardise matrices
    testX = get_standardised_matrix(testX)
    trainX = get_standardised_matrix(trainX)
    
    
    for k in range(3,16):
        algo_label = classify(testX, trainX, trainY, k) #label vorhersagen
        #print(algo_label)
        accuracy = get_accuracy(testY, algo_label) # schauen ob sie übereinstimmen
        #print(accuracy)
        print('For k = ' + str(k) + ' is the accuracy ' + str(accuracy) + '%')

        





testX = 'C:\\BÜSRA\\Uni\\Master\\Essex University\\CE705-7 --- Introduction to Programming in Python\\Assignment\\Data.csv'
testY = 'C:\\BÜSRA\\Uni\\Master\\Essex University\\CE705-7 --- Introduction to Programming in Python\\Assignment\\Data_Labels.csv'

trainX = 'C:\\BÜSRA\\Uni\\Master\\Essex University\\CE705-7 --- Introduction to Programming in Python\\Assignment\\Learning_Data.csv'
trainY = 'C:\\BÜSRA\\Uni\\Master\\Essex University\\CE705-7 --- Introduction to Programming in Python\\Assignment\\Learning_Data_Labels.csv'


run_test(testX, testY, trainX, trainY)
    







```
